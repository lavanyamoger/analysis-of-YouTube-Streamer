```python
import pandas as pd

data = pd.read_csv(r'C:\Users\lavanya\Downloads\youtubers_df.csv')
mean_value = data['Likes'].mean()
median_value = data['Suscribers'].median()
mode_value = data['Rank'].mode()
std_value = data['Comments'].std()

print(f"Mean of the Likes: {mean_value}, Median of the Subscribers: {median_value}, Mode of the rank: {mode_value}, Std Dev of comments: {std_value}")
```
